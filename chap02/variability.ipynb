## 默认配置

开启 HyperThreading 与 TurboBoost

### 情况1：不绑核

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker

execution_times = np.loadtxt("./data/exec_times_parallel.txt")

percentages = (execution_times - execution_times.min()) / execution_times.min()
percentages = np.sort(percentages)

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
fig.tight_layout(pad=3.0)

n_bins = 100

n, bins, patches = axes[0].hist(execution_times, n_bins, density=True, histtype='stepfilled',
                           cumulative=True, label='Empirical')

n, bins, patches = axes[1].hist(execution_times, n_bins, density=False, histtype='bar',
                           cumulative=False, label='Empirical', rwidth=0.8)

axes[2].bar(np.arange(1, 100 + 1), percentages, width=0.6)

axes[0].grid(True)
axes[0].set_xlabel('Exection Time (sec)')
axes[0].set_ylabel('Cumulative Distribution')
axes[1].set_xlim(0.8, 1)

axes[1].grid(True)
axes[1].set_xlabel('Exection Time (sec)')
axes[1].set_ylabel('Density Distribution')
axes[1].set_xlim(0.8, 1)
axes[1].set_ylim(0, 6)

axes[2].grid(True)
axes[2].set_xlabel('Performance Rank of Run')
axes[2].set_ylabel('Percentage Above Mininum')
axes[2].yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals=1))
axes[2].set_ylim(0, 0.12)

fig.suptitle('Distribution of Execution Times for 100 Runs (Parallel)')

plt.show()
```
